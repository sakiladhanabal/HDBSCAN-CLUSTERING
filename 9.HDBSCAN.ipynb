```python
from sklearn.datasets import make_blobs
import hdbscan
```

```python

from sklearn.metrics import silhouette_score, adjusted_rand_score
```

```python
X,y_true=make_blobs(n_samples=1000, centers=5,cluster_std=0.6, random_state=42)
```

```python
clusterer=hdbscan.HDBSCAN(min_cluster_size=5,min_samples=5,cluster_selection_method='eom',allow_single_cluster=True,metric='euclidean',algorithm='best',leaf_size=30)
```

```python
clusterer.fit(X)
```

```python
labels = clusterer.labels_
silhouette_avg = silhouette_score(X, labels)
ari = adjusted_rand_score(y_true, labels)

# Evaluation metrics
print("Silhouette Coefficient: {:.2f}".format(silhouette_avg))
print("Adjusted Rand Index: {:.2f}".format(ari))
```

```python
labels=clusterer.labels_
silhouette_avg = silhouette_score(X,labels)
ari=(adjusted_rand_score(y_true,labels))

print("silhouette Coefficient:{:.2f}".format(silhouette_avg))
print("Adjusted Rand index:{:.2f}".format(ari))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:, 0],X[:,1], c=clusterer.labels_,
            cmap='viridis', s=50,alpha=0.7,edgecolors='k')
plt.colorbar()
plt.title('HDBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
